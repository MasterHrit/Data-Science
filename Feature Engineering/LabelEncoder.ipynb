```python
from sklearn.preprocessing import LabelEncoder
import pandas as pd
import seaborn as sns
import numpy as np
lbl_encoder=LabelEncoder()
```

```python
## Simple DataFrame
df=pd.DataFrame({"color":["red","blue","green","green","red","blue"]})
```

```python
df
```

```python
lbl_encoder.fit_transform(df[["color"]])
```

```python
lbl_encoder.transform([["red"]])
```

```python
## Ordinal Encoding
from sklearn.preprocessing import OrdinalEncoder
```

```python
df=pd.DataFrame({"size":["small","medium","large","medium","small","large"]})
```

```python
df
```

```python
# create instance then fit_transform
o_encoder=OrdinalEncoder(categories=[["small","medium","large"]])
o_encoder.fit_transform(df[["size"]])
```

```python
o_encoder.transform([["small"]])
```
