```python
# SMOTE Minority Oversampling technique

# The Oversampling data getting created will have variance, is not just repetiton of the datapoints at random
# the datapoints are in the line joining the data points
```

```python
from sklearn.datasets import make_classification
```

```python
X,y=make_classification(n_samples=1000,n_features=2,n_clusters_per_class=1,weights=[0.90],random_state=12,n_redundant=0)
```

```python
import pandas as pd
```

```python
df1=pd.DataFrame(X,columns=["f1","f2"])
df2=pd.DataFrame(y,columns=["target"])
```

```python
df_final=pd.concat([df1,df2],axis=1)
```

```python
print(df_final.shape)
print(df_final["target"].value_counts())
print(df_final.tail())
```

```python
import matplotlib.pyplot as plt

plt.scatter(df_final["f1"],df_final["f2"],c=df_final["target"])
```

```python
# to apply smote we use imblearn
from imblearn.over_sampling import SMOTE
```

```python
# Transform the dataset using smote
oversample=SMOTE()
X,y=oversample.fit_resample(df_final[["f1","f2"]],df_final["target"])
```
