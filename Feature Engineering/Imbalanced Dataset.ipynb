```python
# Why handle Imbalanced Dataset
# What is Imbalanced Dataset

'''
Dataset
Trying to solve classification problem
(Supervised ML Problem) -> o/p in form of categories ====> 2 categories (binary classification)

1000 Datapoints
900 yes
100 no

Max are saying yes, the ratio is 9:1, the dataset is imbalanced.
Problem is that the model will get biased towards majority class
we will make both the datapoints equal
1) upsampling - increase the dataset
2) downsampling - decrease the dataset

'''
```

```python
import numpy as np
import pandas as pd

# Set the random seed for reproducibility
np.random.seed(123)

# Create a dataframe with two classes
n_samples=1000
class_0_ratio=0.9
n_class_0=int(n_samples*class_0_ratio)
n_class_1=n_samples-n_class_0
```

```python
n_class_0,n_class_1
```

```python
# Create my dataframe with imbalanced dataset
class_0=pd.DataFrame({
    "feature1":np.random.normal(loc=0,scale=1,size=n_class_0),
    "feature2":np.random.normal(loc=0,scale=1,size=n_class_0),
    "target":[0]*n_class_0
})

class_1=pd.DataFrame({
    "feature1":np.random.normal(loc=0,scale=1,size=n_class_1),
    "feature2":np.random.normal(loc=0,scale=1,size=n_class_1),
    "target":[1]*n_class_1
})
```

```python
df=pd.concat([class_0,class_1]).reset_index(drop=True)
```

```python
df.tail()
```

```python
df["target"].value_counts()
```

```python
df.shape
```

```python
# Upsampling
# Create 2 DataFrames 

df_minority=df[df["target"]==1]
df_majority=df[df["target"]==0]
```

```python
from sklearn.utils import resample

df_minority_upsampled=resample(df_minority,replace=True,
         n_samples=len(df_majority),
         random_state=42) # Sample with replacement
```

```python
df_minority_upsampled.shape
```

```python
df_upsampled=pd.concat([df_majority,df_minority_upsampled])
```

```python
df_upsampled.shape
```

```python
df_upsampled["target"].value_counts()
```

```python
# Create my dataframe with imbalanced dataset
class_0=pd.DataFrame({
    "feature1":np.random.normal(loc=0,scale=1,size=n_class_0),
    "feature2":np.random.normal(loc=0,scale=1,size=n_class_0),
    "target":[0]*n_class_0
})

class_1=pd.DataFrame({
    "feature1":np.random.normal(loc=0,scale=1,size=n_class_1),
    "feature2":np.random.normal(loc=0,scale=1,size=n_class_1),
    "target":[1]*n_class_1
})
```

```python
df=pd.concat([class_0,class_1]).reset_index(drop=True)
```

```python
df["target"].value_counts()
```

```python
## Downsampling

df_minority=df[df["target"]==1]
df_majority=df[df["target"]==0]
```

```python
from sklearn.utils import resample

df_majority_downsampled=resample(df_majority,replace=False,
         n_samples=len(df_minority),
         random_state=42)
```

```python
df_majority_downsampled.shape
```

```python
df_downsampled=pd.concat([df_majority_downsampled,df_minority])
```

```python
df_downsampled.shape
```

```python
df_downsampled["target"].value_counts()

# Here we are losing datapoints (considered bad in most usecases)
```
