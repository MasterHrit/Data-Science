```python
import numpy as np
lst_marks=[45,32,56,72,89,54,32,89,90,87,67,54,45,98,99,67,74]
min,q1,median,q3,max=np.quantile(lst_marks,q=[0,0.25,0.5,0.75,1])
```

```python
min,q1,median,q3,max
```

```python
iqr=q3-q1
iqr
```

```python
lower_fence=q1-(1.5*iqr)
upper_fence=q3+(1.5*iqr)
lower_fence,upper_fence
```

```python
import seaborn as sns
```

```python
sns.boxplot(lst_marks)
```
