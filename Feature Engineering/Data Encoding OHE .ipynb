```python
# OHE - One Hot Encoding --- Disadvantages
# Should not use when we have too many categories, cause it will lead to creation of tooo many features
# Second is Sparse Matrix --- again if there are many categories, the model can lead to overfit a particular feature(created from OHE (since it will have almost 0 and then very less 1))
```

```python
import pandas as pd
from sklearn.preprocessing import OneHotEncoder
```

```python
## Simple DataFrame
df=pd.DataFrame({"color":["red","blue","green","green","red","blue"]})
```

```python
df
```

```python
encoder=OneHotEncoder()
```

```python
# Fit -> Transform (We also have combination of both of them)
encoded=encoder.fit_transform(df[["color"]]).toarray()
# Alphabetically the feature got created blue - green - red
```

```python
import pandas as pd
encoder_df=pd.DataFrame(encoded,columns=encoder.get_feature_names_out())
```

```python
encoder_df
```

```python
# for new data
encoder.transform([["blue"]]).toarray()
```

```python
pd.concat([df,encoder_df],axis=1)
```

```python
import seaborn as sns
```

```python
df=sns.load_dataset("tips")
df
```

```python
df.isnull().sum()
```

```python
ohe=OneHotEncoder()
```

```python
encoded_value=ohe.fit_transform(df[["sex"]]).toarray()
```

```python
sex_df=pd.DataFrame(encoded_value,columns=ohe.get_feature_names_out())
sex_df
```

```python
final_df=pd.concat([df,sex_df],axis=1)
final_df
```

```python
final_df["day"].value_counts()
```

```python
# One Hot Encoding with Day
new_encoder=OneHotEncoder()
```

```python
encoded_values=new_encoder.fit_transform(final_df[["day"]]).toarray()
```

```python
day_encoded_df=pd.DataFrame(encoded_values,columns=new_encoder.get_feature_names_out())
day_encoded_df
```

```python
final_df=pd.concat([final_df,day_encoded_df],axis=1)
```
