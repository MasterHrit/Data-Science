```python
import pandas as pd
```

```python
# Pandas Series - 1D array like object that can hold any datatype
data=[1,2,3,4,5]
series=pd.Series(data)
print(series)
print(type(series))
```

```python
# Create Series from Dict Elements
data={"a":1,"b":2,"c":3}
series_d=pd.Series(data)
print(series_d)
print(type(series_d))
```

```python
data=[10,20,30,40]
index=["A","B","C","D"]
series_i=pd.Series(data,index=index)
print(series_i)
print(type(series_i))
```

```python
# DataFrame - 2D Heterogeneous Tabular Data (Multiple Rows and Columns)
# Creating DataFrame from dictionary of list
d={
    "id":[1,2,3,4],
    "name":["Hritik","Krish","John","Andy"],
    "age":[25,30,35,40]
   }
dataframe=pd.DataFrame(d)
print(dataframe)
print(type(dataframe))
```

```python
import numpy as np
np.array(dataframe)
```

```python
# Create a DataFrame from List of Dictionaries
data=[
    {"id":1,"name":"Hritik","age":27},
    {"id":2,"name":"Krish","age":35},
    {"id":3,"name":"Andrez","age":29}
    ]
df=pd.DataFrame(data)
print(df)
print(type(df))
```
