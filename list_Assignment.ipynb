```python
### Assignment 1: Creating and Accessing Lists

# Create a list of the first 20 positive integers. Print the list.

l=[i for i in range(1,21)]
print(l)
```

```python
# Print the first, middle, and last elements of the list created in Assignment 1.
print("First",l[0])
print("Last",l[-1])
print("Middle",l[len(l)//2])
```

```python
# Print the first five elements, the last five elements, and the elements from index 5 to 15 of the list created in Assignment 1.
print("First Five",l[:5])
print("Last Five",l[-5:])
print("Custom Index",l[5:16])
```

```python
# Create a new list containing the squares of the first 10 positive integers using a list comprehension. Print the new list.
l1=[n**2 for n in range(1,11)]
print(l1)
```

```python
# Create a new list containing only the even numbers from the list created in Assignment 1 using a list comprehension. Print the new list.

l2=[i for i in l if i%2==0]
print(l2)
```

```python
# Create a list of random numbers and sort it in ascending and descending order. Remove the duplicates from the list and print the modified list.
import random
l3=[random.randint(1,100) for i in range(10)]
print(l3)
l3.sort(reverse=False)
print(l3)
l3.sort(reverse=True)
print(l3)

for i in l3:
    if(l3.count(i)>1):
        for j in range(l3.count(i)-1):
            l3.remove(i)
print(l3)
```

```python
# Create a nested list representing a 3x3 matrix and print the matrix. Access and print the element at the second row and third column.

l4=[
    [1,2,3],
    [4,5,6],
    [7,8,9]
    ]
print(l4[1][2])
```

```python
# Create a list of dictionaries where each dictionary represents a student with keys 'name' and 'score'. 
# Sort the list of dictionaries by the 'score' in descending order and print the sorted list.

l5=[{"name":"Hritik","score":95},{"name":"Krish","score":98},{"name":"Karan","score":89},{"name":"Tarun","score":85}]
print(l5)
l5.sort(key=lambda x : x["score"], reverse=True)
print(l5)
```

```python
# Write a function that takes a 3x3 matrix (nested list) as input and returns its transpose. Print the original and transposed matrices.
l4=[
    [1,2,3],
    [4,5,6],
    [7,8,9]
    ]

lt=[]
for i in range(0,3):
    rowlist=[]
    for j in range(0,3):
        rowlist.append(l4[j][i])
    lt.append(rowlist)

print(lt)
```

```python

# Write a function that takes a nested list and flattens it into a single list. Print the original and flattened lists.

print(l4)
fl=[]
for i in l4:
    fl+=i
print(fl)
```

```python
# Create a list of the first 10 positive integers. 
# Remove the elements at indices 2, 4, and 6, and insert the element '99' at index 5. Print the modified list.

l=[i for i in range(1,11)]
print(l)
ld=[val for i,val in enumerate(l) if i not in [2,4,6]]
ld.insert(5,99)

print(ld)
```

```python
# Create two lists of the same length. 
# Use the `zip` function to combine these lists into a list of tuples and print the result.

l1=[1,2,3]
l2=[4,5,6]

nl=[(i,j) for i,j in zip(l1,l2)]
print(nl)
```

```python
# Write a function that takes a list and 
# returns a new list with the elements in reverse order. Print the original and reversed lists.

l=[1,2,3,4]
lr=l[::-1]
print(l,"\n",lr,sep="",end="")
```

```python
### Assignment 14: List Rotation

# Write a function that rotates a list by n positions. Print the original and rotated lists.

l=[1,2,3,4,5]
n=4
rotatelist=l[len(l)-n:]+l[:len(l)-n]
print(l)
print(rotatelist)
```

```python
### Assignment 15: List Intersection

# Write a function that takes two lists and 
# returns a new list containing only the elements that are present in both lists. Print the intersected list.

l1=[1,2,3,4]
l2=[3,4,5,6,7]

il=[i for i in l1 if i in l2]

print(il)
```
