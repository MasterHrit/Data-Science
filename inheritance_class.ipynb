```python
# Inheritance OOP, inherit attr and methods from other class

# Single Inheritance
class Car:
    def __init__(self,doors) -> None:
        self.doors=doors
class Tesla(Car):
    def __init__(self,doors,selfdrive) -> None:
        super().__init__(doors)
        self.selfdrive=selfdrive
```

```python
# Muliple Inheritance ( Class inherits from more than one base class)
#Base Class 1
class Animal:
    def __init__(self,name):
        self.name=name
    def speak(self):
        print("Subclass will implement this")

#Base Class 2
class Pet:
    def __init__(self,owner):
        self.owner=owner

class Dog(Animal,Pet):
    def __init__(self, name,owner):
        Animal.__init__(self,name)
        Pet.__init__(self,owner)
    def speak(self):
        return f"{self.name} says woof"
    
d1=Dog("Tom","MasterHrit")
print(d1.speak())
print(d1.owner)
```
