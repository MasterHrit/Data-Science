```python
## Read a file
with open("example.txt","r") as fo:
    content=fo.read()
    print(content)
```

```python
## Read a file line by line
with open("example.txt","r") as file:
    for line in file:
        print(line.strip()) #.strip operation removes the new line character
```

```python
## Writing a file(Overwriting)
with open("example.txt","w") as file:
    file.write("Hello World!\n")
    file.write("This is a new line.")
```

```python
## Write a file (Without overwriting)
with open("example.txt","a") as file:
    file.write("\nAppend Operation Taking Place\n")
```

```python
# Write Multiple Lines(List of Lines to a file)
lines=["First Line \n","Second Line \n","Third Line \n"]
with open("example.txt","a") as file:
    file.writelines(lines)
```

```python
# Binary Files rb or wb
# Read the content from  a source text file and write to destination text file
# Copying a text file

with open("example.txt","r") as read_file:
    content=read_file.read()
with open("example2.txt","w") as write_file:
    write_file.write(content)
```

```python
## Joining Paths
dir_name="folder"
file_name="file.txt"

full_path=os.path.join(dir_name,file_name)
print(full_path)
```

```python
## Joining Paths
dir_name="folder"
file_name="file.txt"

full_path=os.path.join(os.getcwd(),dir_name,file_name)
print(full_path)
```

```python
path="example.txt"
if os.path.exists(path):
    print(f"The path {path} exists")
else:
    print(f"The path {path} does not exist")
```

```python
# Check if path is file or directory

import os

path="example.txt"

if os.path.isfile(path):
    print("IS FILE")
elif os.path.isdir(path):
    print("ITS DIRECTORY")
```
