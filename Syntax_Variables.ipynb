**Question 17:** Write a Python program to check if a year is a leap year.

```python
# Your code here
y=int(input("Enter the year"))
if(y%100==0):
    if(y%400==0):
        print("Leap")
    else:
        print("Not Leap")
else:
    if(y%4==0):
        print("Leap")
    else:
        print("Not Leap")
```
