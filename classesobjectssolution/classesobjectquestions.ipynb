```python
# Create a class named `MathOperations` with a static method to calculate the square root of a number. Call the static method without creating an object.
import math
class MathOperations:
    @staticmethod
    def square_root(num):
        return math.sqrt(num)

print(MathOperations.square_root(16))
```

abstract class (gives a structural class with necessary features without much implementation (can contain both abstract and normal methods))
abstract method (imp functions necessary to be implemented in derived class)

static method (used @staticmethod decorator) - does not depend on class or instance(self), it is not dependent on object type (it is bounded by class but its usage is independent)
static variable - also called class variable (this does not depend on object, can be accessed through classname or instance)

class method - called by classname, @classmethod decorator used, first arg always as cls(the class itself), it cannot access instance variables

```python
# Create a class named `Rectangle` with private attributes `length` and `width`. 
# Use properties to get and set these attributes. Create an object of the class and test the properties.

class Rectangle:
    def __init__(self,length=0,width=0) -> None:
        self.__length=length
        self.__width=width
    @property
    def length(self):
        return self.__length
    @property
    def width(self):
        return self.__width
    
    @length.setter
    def length(self,value):
        self.__length=value
    @width.setter
    def width(self,value):
        self.__width=value
    

r1=Rectangle()
r1.length=12
r1.width=10
print(r1.length)
print(r1.width)
r1.length=22
print(r1.length)
```

```python
# Create an abstract base class named `Shape` with an abstract method `area`. Create derived classes `Circle` and `Square` that implement the `area` method. 
# Create objects of the derived classes and call the `area` method.
from abc import ABC,abstractmethod
class Shape(ABC):
    @abstractmethod
    def area():
        pass

class Circle(Shape):
    def __init__(self,radius) -> None:
        self.radius=radius
    def area(self):
        return math.pi*(self.radius**2)

class Square(Shape):
    def __init__(self,side) -> None:
        self.side=side
    def area(self):
        return self.side**2
    
c1=Circle(2)
s1=Square(2)
print(c1.area())
print(s1.area())
```

```python
# Create a class named `Vector` with attributes `x` and `y`. 
# Overload the `+` operator to add two `Vector` objects. Create objects of the class and test the operator overloading.
class Vector:
    def __init__(self,x,y) -> None:
        self.x=x
        self.y=y
    def __add__(self, object):
        x=self.x+object.x
        y=self.y+object.y
        return Vector(x,y)
    def __str__(self) -> str:
        return f"Vector({self.x},{self.y})"

v1=Vector(2,3)
v2=Vector(4,5)
print(v1)
print(v2)
print(v1+v2)
```

```python
# Create a custom exception named `InsufficientBalanceError`. 
# In the `BankAccount` class, raise this exception when a withdrawal amount is greater than the balance. Handle the exception and print an appropriate message.
class InsufficientBalance(Exception):
    def __str__(self) -> str:
        return f"Insufficient Balance"
class BankAccount:
    def __init__(self,account_number,balance) -> None:
        self.__account_number=account_number
        self.__balance=balance
    def deposit(self,amount):
        self.__balance+=amount
        print("Deposited",amount)
    def withdraw(self,amount):
        if(amount>self.__balance):
            try:
                raise InsufficientBalance
            except InsufficientBalance as e:
                print(e,f": Withdrawal {amount} is greater than the balance")
        else:
            self.__balance-=amount
            print("Withdrawn",amount)
    def get_balance(self):
        print("Current Balance :",self.__balance)
```

```python
# Create a class named `FileManager` that implements the context manager protocol to open and close a file. Use this class to read the contents of a file.
class FileManager:
    def __init__(self,filename,mode) -> None:
        self.filename=filename
        self.mode=mode
    def __enter__(self):
        self.file=open(self.filename,self.mode)
        return self.file
    def __exit__(self, exc_type, exc, tb):
        self.file.close()
    
with FileManager("sample.txt","r") as file:
    content=file.read()
    print(content)
```

```python
# Create a class named `Calculator` with methods to add, subtract, multiply, and divide. Each method should return the object itself to allow method chaining. Create an object and chain multiple method calls.

class Calculator:
    def __init__(self,value=0) -> None:
        self.value=value
    def add(self,a):
        self.value+=a
        return self
    def sub(self,a):
        self.value-=a
        return self
    def mul(self,a):
        self.value*=a
        return self
    def div(self,a):
        if(a!=0):
            self.value=self.value/a
        else:
            print("Cannot Divide by Zero")
        return self
```

```python
c1=Calculator()
c1.add(5).sub(2).mul(6).div(2)
```
