```python
# ### Assignment 6: Time Series Analysis

# 1. Create a Pandas DataFrame with a datetime index and one column filled with random integers. 
# Resample the DataFrame to compute the monthly mean of the values.
# 2. Create a Pandas DataFrame with a datetime index ranging from '2021-01-01' to '2021-12-31' and 
# one column filled with random integers. Compute the rolling mean with a window of 7 days.

import datetime
df=pd.DataFrame(np.random.randint(1,10,10),index=pd.date_range(start=datetime.date(2026,5,15),end=datetime.date(2026,5,24)))
print(df)

monthly_mean = df.resample("ME").mean()
print(monthly_mean)

rolling_mean=df.rolling(window=7).mean()
print(rolling_mean)
```

```python
# ### Assignment 9: Applying Functions

# 1. Create a Pandas DataFrame with 3 columns and 5 rows filled with random integers. 
# Apply a function that doubles the values of the DataFrame.
# 2. Create a Pandas DataFrame with 3 columns and 6 rows filled with random integers. Apply a lambda function to create a new column that is the sum of the existing columns.

df1=pd.DataFrame(np.random.randint(0,10,15).reshape(5,3))
print(df1)
df1=df1.apply(lambda x:x*2)
print(df1)
```

```python
# # 2. Create a Pandas DataFrame with 3 columns and 6 rows filled with random integers. 
# Apply a lambda function to create a new column that is the sum of the existing columns.

def sumfunc(s):
    sum=0
    for i in s:
        sum+=i
    return sum

df2=pd.DataFrame(np.random.randint(1,10,18).reshape(6,3))
print(df2)
#df2["Sum"]=df2.apply(sumfunc,axis=1)
df2["Sum"]=df2.apply(lambda row:row.sum(),axis=1)
print(df2)
```

```python
# ### Assignment 8: Pivot Tables

# 1. Create a Pandas DataFrame with columns 'Date', 'Category', and 'Value'. 
# Create a pivot table to compute the sum of 'Value' for each 'Category' by 'Date'.
# 2. Create a Pandas DataFrame with columns 'Year', 'Quarter', and 'Revenue'. Create a pivot table to compute the mean 'Revenue' for each 'Quarter' by 'Year'.
import datetime
columns=["Date","category","Value"]
daterng=pd.date_range(start=datetime.date(2026,5,15),end=datetime.date(2026,5,20),freq="D")
l=len(daterng)
df=pd.DataFrame({"Date":np.random.choice(daterng,size=10),"Category":np.random.choice(["A","B","C"],size=10),"Value":np.random.randint(1,10,10)})
print(df)
```

```python
grouped_df=df.groupby("Date")["Value"].sum()
print(grouped_df)
```

```python
pivot_df=df.pivot_table(values="Value",index="Date",columns="Category",aggfunc="sum")
print(pivot_df)
```

```python
# # 2. Create a Pandas DataFrame with columns 'Year', 'Quarter', and 'Revenue'. 
# Create a pivot table to compute the mean 'Revenue' for each 'Quarter' by 'Year'.

df2=pd.DataFrame({"Year":np.random.choice(np.arange(2010,2021),size=10),"Quarter":np.random.choice(["Q1","Q2","Q3","Q4"],size=10),"Revenue":np.random.randint(1,10,10)})
print(df2)

pivot_df2=df2.pivot_table(values="Revenue",index="Year",columns="Quarter",aggfunc="mean")
```
