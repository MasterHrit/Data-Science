```python
## Statistical Concept
## Normalization
# Data change to have mean of 0 and standard dev of 1

data=np.array([1,2,3,4,5])
mean=np.mean(data)
stddev=np.std(data)

normalized_data= (data-mean)/stddev
print("Normalized Data :\n",normalized_data)
```

```python
data=np.arange(1,11)
print(data)
#Mean
print(np.mean(data))
#Median
print(np.median(data))
#Std
print(np.std(data))
#Variance
print(np.var(data))
```

```python
## Logical Operations
data=np.arange(1,11)
print(data)
print(data>5)
data[(data>=5) & (data<=8)]
```

```python
df_filled_na=df.fillna(0)
```

```python
### Filling Missing Values with the Mean of the column
df["Sales_FilledNA"]=df["Sales"].fillna(df["Sales"].mean())
```

```python
df.head()
```

```python
# Rename columns
# Date --> Sales Date
df=df.rename(columns={"Date":"Sale Date"})
print(df.head())
```

```python
# Change Data types
# fillna Value with mean then
# Value --> Int
df.isna().sum(axis=0)

df["Value_int"]=df["Value"].fillna(df["Value"].mean()).astype(int)
df.head(20)
```

```python
# New Col : 2*value
# use lambda and apply function on series
df["Value_Double"]=df["Value_int"].apply(lambda x:x*2)
df.head(20)
```

```python
# Data Agg and Grouping
#Group by Product and get the mean

grouped_prod_mean=df.groupby("Product")["Value"].mean()
grouped_prod_mean
#type(grouped_prod_mean)
```

```python
grouped_prod_reg_mean=df.groupby(["Product","Region"])["Value"].mean()
```

```python
grouped_prod_reg_mean
```

```python
grouped_agg=df.groupby("Region")["Value"].agg(["mean","max","median","count","sum"])
print(grouped_agg)
```

```python
# Merging and Joining Dataframes
df1=pd.DataFrame({"Key":["A","B","C"],"Value1":[1,2,3]})
df2=pd.DataFrame({"Key":["A","B","D"],"Value2":[2,4,6]})

df_merged=pd.merge(df1,df2,on="Key",how="inner")
print(df_merged)
```

```python
df_merged=pd.merge(df1,df2,on="Key",how="left")
df_merged
```

```python
df_merged=pd.merge(df1,df2,on="Key",how="right")
df_merged
```
