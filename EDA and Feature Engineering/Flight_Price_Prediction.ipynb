```python
def splithour(s):
    split=s.split(" ")
    if(len(split)>=2 or "h" in split[0]):
        return int(split[0].split("h")[0])
    else:
        return 0
```

```python
splithour("10h 50m")
```

```python
def splitminute(s):
    split=s.split(" ")
    if(len(split)<2):
        if("m" in split[0]):
            return int(split[0].split("m")[0])
        else:
            return 0
    else:
        return int(split[1].split("m")[0])
```

```python
splitminute("10h 5m")
```

```python
df["Duration_hour"]=df["Duration"].apply(splithour)
```

```python
df["Duration_minute"]=df["Duration"].apply(splitminute)
```
